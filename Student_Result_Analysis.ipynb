**Drop Unnamed Column**

```python
df = df.drop('Unnamed: 0', axis=1)
print(df.head())
```

**Change Weekly study hours column**

```python
df['WklyStudyHours'] = df['WklyStudyHours'].str.replace('05-Oct','5-10')
df.head()
```

**gender distribution**

```python
plt.figure(figsize=(6,5))
ax = sns.countplot(data=df, x='Gender', hue='Gender', palette=['orange', 'skyblue'], legend=False)
for container in ax.containers:
    ax.bar_label(container)
plt.title('Gender Distribution')

plt.show()
```

from the above chat we have analysed that the number of females is more than the number of male

```python
gb = df.groupby('ParentEduc').agg({'MathScore': 'mean', 'ReadingScore': 'mean', 'WritingScore': 'mean'})
print(gb)
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(gb, annot = True)
plt.title("Relationship between parent's Education and student's Score")
plt.show()
```

From the above chart we have concluded that the education of parents have a good impact on your Scores

```python
gb1= df.groupby('ParentMaritalStatus').agg({'MathScore': 'mean', 'ReadingScore': 'mean', 'WritingScore': 'mean'})
print(gb1)
```

```python
sns.heatmap(gb1, annot = True)
plt.title("Relationship between parent's marital status and student's Score")
plt.show()
```

from the above chart we have calculated that there is no/negligible impact on the on the student's score 
due to their parent's marital status

**Distribution of Ethninic Groups**

```python
groupA = df.loc[(df['EthnicGroup'] == 'group A')].count()
groupB = df.loc[(df['EthnicGroup'] == 'group B')].count()
groupC = df.loc[(df['EthnicGroup'] == 'group C')].count()
groupD = df.loc[(df['EthnicGroup'] == 'group D')].count()
groupE = df.loc[(df['EthnicGroup'] == 'group E')].count()
print(groupA,groupB,groupC,groupD,groupE)
```

```python
l=['groupA','groupB','groupC','groupD','groupE']
mlist = [groupA['EthnicGroup'],groupB['EthnicGroup'],groupC['EthnicGroup'],groupD['EthnicGroup'],groupE['EthnicGroup']]
plt.figure(figsize=(7,9))
plt.title('Distribution of EthnicGroups')
plt.pie(mlist,labels=l,autopct= '%1.2f%%')
plt.show()
print(mlist)
```
